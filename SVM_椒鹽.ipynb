```python
data_path = '/content/gdrive/Shareddrives/AI Final Project/dataset'
```

```python
import pandas as pd
import os
from skimage.transform import resize
from skimage.io import imread
import numpy as np
import matplotlib.pyplot as plt
import joblib
from thundersvm import SVC
from cv2 import medianBlur
```

## **Dataset, Data Loader, and Transforms**

```python
Categories=['freshapples','freshbanana','freshoranges','rottenapples','rottenbanana','rottenoranges']
#Categories=['freshapples','rottenapples']
```

```python
flat_data = [] #input array
target = [] #output array

data_path_train = data_path + "/train"

#path which contains all the categories of images
for i in Categories:
  print(f'loading... category : {i}')
  path = os.path.join(data_path_train,i)
  count = 0
  count_sp = 0
  for img in os.listdir(path):
    count += 1
    if count%100 == 0:
      print(count)
    img_array = imread(os.path.join(path,img))
    if "saltandpepper" in img:
      img_array = medianBlur(img_array, 3) # 去除椒鹽
      count_sp+=1
    img_resized = resize(img_array,(64,64,3))
    flat_data.append(img_resized.flatten())
    target.append(Categories.index(i))
  print("salt and pepper: ",count_sp)
  print(f'loaded category:{i} successfully')
flat_data = np.array(flat_data)
target = np.array(target)
```

## **Model**


```python
# Initialize model
model = SVC(C=500, kernel='rbf')
```

## **Training**

```python
import time
t1 = time.time()

model.fit(flat_data, target)
print('The Model is trained well with the given images')
# model.best_params_ contains the best parameters obtained from GridSearchCV

t2 = time.time()
print('time elapsed: ' + str(t2-t1) + ' seconds')

#save your model or results
joblib.dump(model, 'model_svm_64_sp.pkl')
```

```python
############### 只留有椒鹽的 #######################

#要記得去前面檢查現在category有哪些

flat_data_test = [] #input array
target_test = [] #output array

data_path_test = data_path+"/test"

#path which contains all the categories of images
for i in Categories:
  print(f'loading... category : {i}')
  path = os.path.join(data_path_test,i)
  count = 0
  for img in os.listdir(path):
    if "saltandpepper" not in img:
      continue
    count += 1
    if count%100 == 0:
      print(count)
    img_array = imread(os.path.join(path,img))
    #img_array = medianBlur(img_array, 3) # 去除椒鹽
    img_resized = resize(img_array,(64,64,3))
    flat_data_test.append(img_resized.flatten())
    target_test.append(Categories.index(i))
  print(f'loaded category:{i} successfully')
flat_data_test = np.array(flat_data_test)
target_test = np.array(target_test)
```

```python
from sklearn.metrics import accuracy_score
from sklearn.metrics import classification_report
from sklearn.metrics import confusion_matrix
```

```python
print("Test data")

y_pred = model.predict(flat_data_test)
print(f"The model is {accuracy_score(target_test, y_pred)*100}% accurate")
print("report:\n",classification_report(target_test, y_pred, target_names=Categories, digits=4))
print("confusion matrix:\n",confusion_matrix(target_test, y_pred))
```
