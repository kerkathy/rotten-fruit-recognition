```python
import pandas as pd
import os
from skimage.transform import resize
from skimage.io import imread
import numpy as np
import matplotlib.pyplot as plt
import joblib
```

```python
Categories=['freshapples','freshbanana','freshoranges','rottenapples','rottenbanana','rottenoranges']
#Categories=['freshapples','rottenapples']
```

#####Print all resized img

```python
data_path_test = data_path + "/test"

fig = plt.figure(figsize=(18, 28))
columns = 4
rows = 9

# ax enables access to manipulate each of subplots
ax = []

print('Resized image:')
count = 0

for i in Categories:
  path = os.path.join(data_path_test, i)
  for img in os.listdir(path):
    img_array = imread(os.path.join(path,img))
    img_resized = resize(img_array,(64,64,3))

    # create subplot and append to ax
    ax.append( fig.add_subplot(rows, columns, count+1) )
    ax[-1].set_title('{}'.format(img))  # set title
    plt.imshow(img_resized)
    count += 1
    if count % 4 == 0:
      break

plt.show()
```
